file

flush
  whether to forcibly flush the stream.

```python
# static typing (c/c++/java/c#) -> in this programming we know pre-define data type is avalible. and alwyas we have define a data type.
# so, what happened... in same variable different type value do not assign. if we have assign similar type of value than we can get an error

# example: in c++... that's why it's called static typing
# int num = 10;
# num = "kavit" // we have get error. conflicting data type.

# in python... dynamic typing
# example

a = 10
print(a)
print()

# dynamic binding
a = 10
print(a)
a = "kavit"
print(a)
a = True
print(a)
```
